```python
import os
import pandas as pd

# Lista de encabezados
headers = ["Código centro","Centro","Ubicac. técnica","Denominación Ubic.","Equipo","Denominación","No. Plan","Desc. Plan","Orden","Actividad","Clase de orden","Status sistema","Inicio program.","Responsable","Prioridad","empty","Duración"]

# Carpeta que contiene los archivos CSV
carpeta = r"..\No_programados copy\dataset_actividades_no_programadas"

# DataFrame vacío para almacenar todos los datos

def merg(carpeta,headers,name):
    df_combinado = pd.DataFrame()
    # Iterar sobre todos los archivos en la carpeta
    for filename in os.listdir(carpeta):
        if filename.endswith('.csv'):
            archivo_path = os.path.join(carpeta, filename)
            try:
                # Leer el archivo CSV y especificar los encabezados
                df = pd.read_csv(archivo_path, header=1, names=headers, na_values=['NaN', 'NA'], dtype=str)

                # Concatenar el DataFrame actual al DataFrame combinad
                df_combinado = pd.concat([df_combinado, df], ignore_index=True)
            except:
                pass
    try:
        # Guardar el DataFrame combinado en un archivo CSV
        umbral_nulos = len(df.columns) * 0.8  # Define el umbral como el 90% de las columnas
        df_limpiado = df_combinado.dropna(thresh=umbral_nulos)
        df_limpiado.to_csv(f'{name}_clean_.csv', index=False)
    except:
        pass
```

```python
merg(r"..\No_programados copy\dataset_inspecciones_no_programadas", headers=headers,name="inspecciones_no_programadas")
```
