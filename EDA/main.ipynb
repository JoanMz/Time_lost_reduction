```python
import os
from openpyxl import load_workbook
import csv

def save_sheet_as_csv(file_path, sheet_name, output_folder):
    """
    Saves a specific sheet from an .xlsx file as a CSV file in a specified output path.
    The CSV file name will be a combination of the containing folder name, the original file name, and the sheet name.
    """
    # Load the Excel workbook and the specific sheet
    workbook = load_workbook(file_path)
    sheet = workbook[sheet_name]
    
   # Construct the output file name
    folder_name = os.path.basename(os.path.dirname(file_path))
    base_name = os.path.splitext(os.path.basename(file_path))[0]  # Nombre del archivo sin extensión
    output_file_name = f"{base_name}_{folder_name}_ACTIVIDADES.csv"
    output_path = os.path.join(output_folder, output_file_name)
    
    # Write the sheet data to a CSV file
    with open(output_path, 'w', newline="", encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        for row in sheet.iter_rows(values_only=True):
            writer.writerow(row)

def process_xlsx_files(file_path, output_folder):
    """
    Processes .xlsx files to find and convert sheets named "ACTIVITIES" into CSV files.
    """
    try:
        workbook = load_workbook(file_path, read_only=True)
        if "ACTIVIDADES" in workbook.sheetnames:
            save_sheet_as_csv(file_path, "ACTIVIDADES", output_folder)
    except Exception as e:
        print(f"Error al procesar el archivo {file_path}: {e}")

def list_files_and_directories(root_path, output_folder):
    """
    Walks through directories and files from a root path, looking for .xlsx files to process.
    """
    for path, directories, files in os.walk(root_path):
        for file in files:
            if file.endswith(".xlsx"):
                file_path = os.path.join(path, file)
                process_xlsx_files(file_path, output_folder)

output_folder = "C:\\Users\\cesar\\Desktop\\platzi\\Time_lost_reduction\\dataset_actividades_no_programadas"
paths = [
    "C:\\Users\\cesar\\Desktop\\BIBLIOCASTIA\\NO_progrmados_code"
]

for root_path in paths:
    list_files_and_directories(root_path, output_folder)
```
