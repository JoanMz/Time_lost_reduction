```python
import pandas as pd
import os
import re
```

```python
def procesar_archivo_csv(archivo):
    # Lee el archivo CSV especificando el tipo de datos de cada columna
    df = pd.read_csv(archivo, dtype=str)

    # Llena los valores NaN con una cadena vacía
    df = df.fillna('')

    # Encuentra la fila que contiene los nombres de las columnas utilizando expresiones regulares
    columnas_row = df[df.iloc[:, 0].str.contains("Código centro")]

    # Verifica si se encontró al menos una fila que coincida con el patrón
    if not columnas_row.empty:
        # Encuentra el índice de la primera fila que contiene los nombres de las columnas
        columnas_index = columnas_row.index[0]

        # Elimina las filas anteriores a la fila de los nombres de las columnas
        df = df.iloc[columnas_index:]

        # Guarda el DataFrame modificado en el mismo archivo
        df.to_csv(archivo, index=False)
    else:
        print(f"No se encontraron nombres de columnas en el archivo: {archivo}")
        os.remove(archivo)
        print("rm")
```

```python
# Directorio donde se encuentran los archivos CSV
directorio_No_programados_inspecciones = r"..\No_programados copy\dataset_inspecciones_no_programadas"
directorio_No_programados_actividades =  r"..\No_programados copy\dataset_actividades_no_programadas"
directorio_programados_inspecciones = r"..\programados copy\dataset_inspecciones_programadas"
directorio_programados_actividades = r"..\programados copy\dataset_actividades_programadas"
```

```python
# Itera sobre todos los archivos CSV en el directorio
for filename in os.listdir(directorio_No_programados_inspecciones):
    if filename.endswith('.csv'):
        archivo_path = os.path.join(directorio_No_programados_inspecciones, filename)
        procesar_archivo_csv(archivo_path)
```

```python
import os
import re

def procesar_archivo_csv(archivo):
    # Abre el archivo CSV en modo de lectura
    with open(archivo, 'r', encoding='utf-8') as f:
        # Lee las dos primeras líneas del archivo
        lineas = [f.readline().strip() for _ in range(2)]

    # Verifica si hay al menos dos líneas en el archivo
    if len(lineas) == 2:
        # Divide cada línea en una lista de columnas
        encabezado1 = lineas[0].split(',')
        encabezado2 = lineas[1].split(',')

        # Combina las dos listas de columnas para formar un nuevo encabezado
        nuevo_encabezado = ','.join(encabezado1[:-1] + encabezado2)

        # Busca el patrón que coincide con "Código centro" en el nuevo encabezado
        if re.search(r"Código centro", nuevo_encabezado):
            # Crea un nombre de archivo nuevo para escribir el archivo modificado
            nuevo_archivo = archivo + '.modificado'

            # Escribe el nuevo encabezado en el nuevo archivo
            with open(nuevo_archivo, 'w', encoding='utf-8') as f_nuevo:
                f_nuevo.write(nuevo_encabezado + '\n')
                
                # Reabre el archivo original en modo de lectura
                with open(archivo, 'r', encoding='utf-8') as f_original:
                    # Salta las dos primeras líneas
                    next(f_original)
                    
                    # Copia las líneas restantes al nuevo archivo
                    for linea in f_original:
                        f_nuevo.write(linea)
            
            # Elimina el archivo original
            os.remove(archivo)
            print(f"Se eliminó el archivo original: {archivo}")

            # Renombra el nuevo archivo al nombre original
            os.rename(nuevo_archivo, archivo)
            print(f"Se renombró el archivo modificado: {nuevo_archivo} -> {archivo}")
        else:
            # Elimina el archivo si no se encuentra el patrón
            os.remove(archivo)
            print(f"Se eliminó el archivo debido a que no se encontró el patrón: {archivo}")
    else:
        # Elimina el archivo si tiene menos de dos líneas
        os.remove(archivo)
        print(f"Se eliminó el archivo debido a que tiene menos de dos líneas: {archivo}")

# Directorio donde se encuentran los archivos CSV
directorio_No_programados_inspecciones = r"..\No_programados copy\dataset_inspecciones_no_programadas"
directorio_No_programados_actividades = r"..\No_programados copy\dataset_actividades_no_programadas"
directorio_programados_inspecciones = r"..\programados copy\dataset_inspecciones_programadas"
directorio_programados_actividades = r"..\programados copy\dataset_actividades_programadas"

# Itera sobre todos los archivos CSV en los directorios
directorios = [
    directorio_No_programados_inspecciones,
    directorio_No_programados_actividades,
    directorio_programados_inspecciones,
    directorio_programados_actividades
]

for directorio in directorios:
    for filename in os.listdir(directorio):
        if filename.endswith('.csv'):
            archivo_path = os.path.join(directorio, filename)
            procesar_archivo_csv(archivo_path)
```
