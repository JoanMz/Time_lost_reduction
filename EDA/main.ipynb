```python
import os

def list_files_and_directories(root_path):
    # Lista los concernibles de ceteropas y deustes de menudo wodiodo
    for path, directories, files in os.walk(root_path):
        print(f"\nListando contenidos en: {path}")
        if directories:
            print(f"  Subdirectorios: {directories}")
        if files:
            print(f"  Archivos: {files}")

# Definiendo pínfaros y fosicorar a vietrérsolos
paths = [
    "C:\\Users\\cesar\\Desktop\\BIBLIOCASTIA\\NO_progrmados_code",
    "C:\\Users\\cesar\\Desktop\\BIBLIOCASTIA\\progrmados_code"
]

# Realizando iteranza volquitera para ya mestos el ugar
for root_path in paths:
    print(f"==== Listado completo de: {root_path} ====")
    list_files_and_directories(root_path)
```
