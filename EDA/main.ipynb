```python
import os
from openpyxl import load_workbook

def remove_spaces_in_sheet_names(file_path):
    # Carga el libro de trabajo
    workbook = load_workbook(file_path)
    modified = False
    
    # Itera sobre los nombres de las hojas y los modifica si es necesario
    for sheet in workbook.sheetnames:
        new_name = sheet.strip()
        if new_name != sheet:
            modified = True
            workbook[sheet].title = new_name
    
    # Guarda el libro si se modificaron los nombres de las hojas
    if modified:
        workbook.save(file_path)

def list_files_and_directories(root_path):
    # Listar el contenido de los directorios y archivos de manera considerada
    for path, directories, files in os.walk(root_path):
        for file in files:
            # Asumiendo que quieres trabajar con archivos .xlsx
            if file.endswith(".xlsx"):
                file_path = os.path.join(path, file)
                remove_spaces_in_sheet_names(file_path)

# Definir las rutas de búsqueda y proceder a revelarlas
paths = [
    "C:\\Users\\cesar\\Desktop\\BIBLIOCASTIA\\NO_progrmados_code",
    "C:\\Users\\cesar\\Desktop\\BIBLIOCASTIA\\progrmados_code"
]

# Realizar iterancia sobre las rutas especificadas
for root_path in paths:
    list_files_and_directories(root_path)
```

```python
import os
from openpyxl import load_workbook
import csv

def save_sheet_as_csv(file_path, sheet_name, output_folder):
    """
    Guarda una hoja específica de un archivo .xlsx como un archivo CSV en una ruta de salida especificada.
    El nombre del archivo CSV será una combinación del nombre de la carpeta contenedora, el nombre del archivo original y el nombre de la hoja.
    """
    # Cargar el libro de Excel y la hoja específica
    workbook = load_workbook(file_path)
    sheet = workbook[sheet_name]
    
    # Construir el nombre del archivo de salida
    folder_name = os.path.basename(os.path.dirname(file_path))
    base_name = os.path.splitext(os.path.basename(file_path))[0]  # Nombre del archivo sin extensión
    output_file_name = f"{base_name}_{folder_name}_ACTIVIDADES.csv"
    output_path = os.path.join(output_folder, output_file_name)
    
    # Escribir los datos de la hoja a un archivo CSV
    with open(output_path, 'w', newline="", encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        for row in sheet.iter_rows(values_only=True):
            writer.writerow(row)

def process_xlsx_files(file_path, output_folder):
    """
    Procesa archivos .xlsx para encontrar y convertir hojas llamadas "ACTIVIDADES" en archivos CSV.
    """
    try:
        workbook = load_workbook(file_path, read_only=True)
        if "ACTIVIDADES" in workbook.sheetnames:
            save_sheet_as_csv(file_path, "ACTIVIDADES", output_folder)
    except Exception as e:
        print(f"Error al procesar el archivo {file_path}: {e}")

def list_files_and_directories(root_path, output_folder):
    """
    Recorre los directorios y archivos a partir de una ruta raíz, buscando archivos .xlsx para procesar.
    """
    for path, directories, files in os.walk(root_path):
        for file in files:
            if file.endswith(".xlsx"):
                file_path = os.path.join(path, file)
                process_xlsx_files(file_path, output_folder)

output_folder = "C:\\Users\\cesar\\Desktop\\platzi\\Time_lost_reduction\\dataset_actividades_no_programadas"
paths = [
    "C:\\Users\\cesar\\Desktop\\BIBLIOCASTIA\\NO_progrmados_code"
]

for root_path in paths:
    list_files_and_directories(root_path, output_folder)
```
